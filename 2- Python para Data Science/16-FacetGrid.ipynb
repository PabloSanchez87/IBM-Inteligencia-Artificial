```python
import pandas as pd
import os

from matplotlib import pyplot as plt
import seaborn as sns
%matplotlib inline
import warnings
warnings.filterwarnings('ignore')

peliculas = pd.read_csv('RatingsPeliculas.csv')
peliculas.columns = ['Pelicula', 'Genero', 'RatingsCriticos', 'RatingsAudiencia', 'PresupuestoMillones', 'A単o']

peliculas.Pelicula = peliculas.Pelicula.astype('category')
peliculas.Genero = peliculas.Genero.astype('category')
peliculas.A単o = peliculas.A単o.astype('category')
peliculas.info()
```

---

# FacetGrid

```python
# Creando un FacetGrid

g = sns.FacetGrid(data=peliculas, row="Genero", col='A単o', hue="Genero")
g = g.map(plt.scatter, 'RatingsCriticos', 'RatingsAudiencia')
```

```python
# Creando un FacetGrid

g = sns.FacetGrid(data=peliculas, row="Genero", col='A単o', hue="Genero")
g = g.map(plt.hist, "PresupuestoMillones")
```

```python
g = sns.FacetGrid(data=peliculas, row="Genero", col='A単o', hue="Genero")

kws = dict(s=50, edgecolor='black', linewidth=0.5) 

g = g.map(plt.scatter, 'RatingsCriticos', 'RatingsAudiencia', **kws)
```

# Cordinadas y diagonales

```python
# Agregando ejes y agregando diagonales

g = sns.FacetGrid(data=peliculas, row="Genero", col='A単o', hue="Genero")
kws = dict(s=50, edgecolor='black', linewidth=0.5) 
g = g.map(plt.scatter, 'RatingsCriticos', 'RatingsAudiencia', **kws)
g.set(xlim=(0, 100), ylim=(0, 100))

for ax in g.axes.flat:
    ax.plot((0,100), (0,100), c='gray', ls='--')
    ax.set(xlabel='RatingsCriticos', ylabel='RatingsAudiencia')
    
g.add_legend()
```
