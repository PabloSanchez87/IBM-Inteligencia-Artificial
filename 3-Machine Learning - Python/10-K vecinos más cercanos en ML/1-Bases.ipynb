# K vecinos ms cercanos - kNN
- El mtodo de los k vecinos ms cercanos es un mtodo de clasificacin supervisada, que estima la probabilidad de que un elemento 'x' pertenezca a una clase 'C' a partir de la informacin proporcionada.
- En el reconocimiento de patrones, este algoritmo es usado como mtodo de clasificacin de objetos, basado en un entrenamiento mediante ejemplo cercanos en el espacio de los elementos.
- Es sencillo, fcil de entrenar y trabaja con cualquier mmero de clases.

    

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dataframe = pd.read_csv('vehiculos.csv')
dataframe.head()
```

```python
dataframe.columns
```

```python
dataframe['vehicle_class'].unique()
```

```python
y = dataframe['vehicle_class']
X = dataframe.drop('vehicle_class', axis=1)
```

```python
y
```

```python
X
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=45)
```

```python
X_train
```

```python
from sklearn.neighbors import KNeighborsClassifier

knn = KNeighborsClassifier(n_neighbors=1)
```

```python
knn.fit(X_train, y_train)
```

```python
predicciones = knn.predict(X_test)
predicciones
```

```python
from sklearn.metrics import classification_report, confusion_matrix
print(confusion_matrix(y_test, predicciones))
```

```python
print(classification_report(y_test, predicciones))
```

```python
# Buscamos el valor de K que maximiza el error de regresin

tasa_error = []
for i in range(1, 30):
    knn = KNeighborsClassifier(n_neighbors=i)
    knn.fit(X_train, y_train)
    prediccion_i = knn.predict(X_test)
    tasa_error.append(np.mean(y_test != prediccion_i))
```

```python
valores = range(1, 30)
plt.plot(valores, tasa_error, color='g', marker='o', markersize=8, markerfacecolor='red')
```

```python
knn = KNeighborsClassifier(n_neighbors=3)
knn.fit(X_train, y_train)
predicciones = knn.predict(X_test)
```

```python
print(confusion_matrix(y_test, predicciones))
```

```python
print(classification_report(y_test, predicciones))
```
