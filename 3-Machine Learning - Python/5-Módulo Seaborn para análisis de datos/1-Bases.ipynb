## Mapas de calor (heatmaps)

```python
vuelos = sns.load_dataset('flights')
vuelos.head(10)
```

```python
vuelos_matrix = vuelos.pivot_table(index='month', columns='year', values='passengers', observed=False)

vuelos_matrix.head()
```

```python
sns.heatmap(vuelos_matrix, cmap='coolwarm', cbar=False)
```

```python
sns.heatmap(vuelos_matrix, cmap='coolwarm', linecolor='white',  linewidths=0.5)
```

## Gráficos de cuadrícula (grids): PairGrid y FacetGrid

```python
import matplotlib.pyplot as plt
flores = sns.load_dataset('iris')
flores.head()
```

```python
sns.pairplot(flores)
```

```python
graficos = sns.PairGrid(flores)
graficos.map(plt.scatter)
```

```python
graficos = sns.PairGrid(flores)
graficos.map_diag(sns.kdeplot)
graficos.map_upper(plt.scatter)
graficos.map_lower(sns.kdeplot)
```

## Gráficos de regresión(Regression Plots)

```python
propinas = sns.load_dataset('tips')
propinas.head()
```

```python
sns.lmplot(x='total_bill', y='tip', data=propinas, hue='day')
```

```python
sns.lmplot(x='total_bill', y='tip', data=propinas, hue='sex', fit_reg=False, markers=['o', 'x'])
```

```python
sns.lmplot(x='total_bill', y='tip', data=propinas, hue='sex', fit_reg=False, markers=['o', 'x'], scatter_kws={'s': 80})
```

```python
sns.lmplot(x='total_bill', y='tip', data=propinas)
```

```python
sns.lmplot(x='total_bill', y='tip', data=propinas, col='sex')
```

```python
sns.lmplot(x='total_bill', y='tip', data=propinas, col='day')
```

```python
sns.lmplot(x='total_bill', y='tip', data=propinas, col='day', row='sex')
```

```python
sns.lmplot(x='total_bill', y='tip', data=propinas, col='day', row='sex', hue='smoker')
```

```python
sns.lmplot(x='total_bill', y='tip', data=propinas, aspect=0.8)
```
