# Matplotlib

- Bibliotea para la generación de gáficos a partir de datos contenidos en listas o arrays en
- Control sobre cada aspecto de la visualización

- https://matplotlib.org/stable/gallery/

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(2, 9, 10)
x
```

```python
y = x**2
y
```

```python
plt.plot(x, y,'red')
plt.title('Title')
plt.xlabel('axis x')
plt.ylabel('axis y')
plt.show()
```

```python
plt.subplot(1,2,1) # 1 column, 2 rows
plt.plot(x,y,'green')
plt.subplot(1,2,2) # 1 column, 2 rows
plt.plot(y,x,'blue')
```

```python
plt.subplot(2,1,1) # 2 columnas, 1 fila
plt.plot(x,y,'green')
plt.subplot(2,1,2) # 2 columnas, 1 fila
plt.plot(y,x,'blue')
```

## Aplicamos POO a la visualización de datos

```python
figura = plt.figure()
grafico = figura.add_axes([0.1, 0.1, 0.8, 0.8]) # Definimos el área de la figura, en este caso un cuadro de 0.1 por 0.1 de ancho y alto (0.8 por 0.8)
grafico.plot(x, y, 'green')
```

```python
figura = plt.figure()
grafico1 = figura.add_axes([0.1, 0.1, 0.8, 0.8]) # posición y anchura de la figura en el espacio de coordenadas de la figura (0,0) es la esquina superior izquierda de la figura y (1,1) es la esquina inferior derecha de la figura 
grafico1.plot(x, y)

grafico2 = figura.add_axes([0.2, 0.5, 0.3, 0.3]) # posición y anchura
grafico2.plot(x, y, 'g')
```

## Multigráficos con Matplotlib

```python
figura, graficos = plt.subplots(nrows=2, ncols=2)
```

```python
figura, graficos = plt.subplots(nrows=2, ncols=2)

graficos[0][0].plot(x, y, 'b')
graficos[0][0].set_title('Gráfico 1')

graficos[0][1].plot(x, y, 'r')
graficos[0][1].set_title('Gráfico 2')

graficos[1][0].plot(x, y, 'g')
graficos[1][0].set_title('Gráfico 3')

graficos[1][1].plot(x, y, 'y')
graficos[1][1].set_title('Gráfico 4')
```

## Tamaño del gráfico

```python
figura = plt.figure(figsize=(10,5))
grafico = figura.add_axes([0.1,0.1,0.8,0.8])
grafico.plot(x,y, 'r') 
```

## Dos gráficos en la misma figura

```python
figura = plt.figure()

grafico = figura.add_axes([0.1, 0.1, 0.8, 0.8])
grafico.plot(x, x**2, 'r', label='x^2')
grafico.plot(x, x**3, 'b', label='x^3')
grafico.legend(loc=0) # loc=0 para ubicar la leyenda en la esquina superior izquierda
```

## Color del gráfico

```python
figura = plt.figure()
grafico = figura.add_axes([0, 0, 1, 1])
grafico.plot(x, y, '#a4ccee')
```

## Tipo de gráfico (grosor, transparencia, etc.)

```python
figura = plt.figure()
grafico = figura.add_axes([0, 0, 1, 1])

grafico.plot(x, y, 'r', linewidth=4, alpha=0.5, linestyle='-.')
```
