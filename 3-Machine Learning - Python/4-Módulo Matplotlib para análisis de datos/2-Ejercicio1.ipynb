- Crear una variable x que tenga 10 valores equidistantes entre los valores 0 y 20
- Crear una variable y que tenga los valores de la funcion x al cuadrado + 5
- Crear un gr√°fico de color rojo, con ancho 5 y estilo de puntos

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0, 20, 10)
x
```

```python
y = (x**2)+5
y
```

```python
plt.plot(x, y, 'r',linestyle=':', linewidth=5)
```
