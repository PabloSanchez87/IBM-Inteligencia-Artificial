# rboles de decisin
- Un rbol de decisin es un modelo de prediccin utilizado en diversos mbitos que van desde la inteligencia artificial hasta la economa.
- Dado un conjuntos de datos, se fabrican diagramas de construcciones lgicas, que sirven para representar y categorizar una serie de condiciones que ocurren de forma sucesiva, para la resolicin de un problema.
- Los rboles de decisin se pueden clasificar en tres tipos:
    - rboles de decisin binarios: se utilizan para clasificar datos binarios.
    - rboles de decisin mltiples: se utilizan para clasificar datos que pueden tomar mltiples valores.
    - rboles de decisin no lineales: se utilizan para clasificar datos que no siguen una lnea lineal.

    

# rbol de decisin - clasificacin

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
vinos = pd.read_csv('vino.csv')
vinos.head()
```

```python
vinos['Wine Type'].unique()
```

```python
vinos['Wine Type'].value_counts()
```

```python
X = vinos.drop(['Wine Type'] , axis=1)
y = vinos['Wine Type']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)
```

```python
X_train
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
arbol = DecisionTreeClassifier()
```

```python
arbol.fit(X_train, y_train)
```

```python
predicciones = arbol.predict(X_test)
predicciones
```

```python
y_test
```

```python
from sklearn.metrics import confusion_matrix, classification_report
```

```python
print(classification_report(y_test, predicciones))
```

```python
print(confusion_matrix(y_test,predicciones))
```
