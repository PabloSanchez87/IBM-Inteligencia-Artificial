# Plotly and Cunfflinks

- Plotly es una libería de Python que permite crear gráficos interactivos y presentarlos en línea.
    - 'pip install plotly'
- Cunflink es una librería de Python que ayuda a conectar Plotly con Pandas.
    - 'pip install cufflinks'

```python
import pandas as pd
import numpy as np

import cufflinks as cf
from plotly.offline import init_notebook_mode, iplot, plot, download_plotlyjs
```

```python
init_notebook_mode(connected=True) # <-- This is the magic line
```

```python
cf.set_config_file(offline=True)  # Configurar `cufflinks` en modo offline
cf.go_offline()  # Para asegurar visualización offline
cf.set_config_file(world_readable=True, theme='pearl')  # Configuración adicional opcional
```

```python
%matplotlib inline 
# Para mostrar gráficos en la consola
```
