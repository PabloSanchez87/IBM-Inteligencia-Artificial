## Tipos de algoritmos de Machine Learning
- `Aprendizaje supervisado`: Este algoritmo necesita datos prviamente etiquetados para aprender a realizar el trabajo. En base a estos datos, el algoritmo es capaz de aprender a resolver problemas futuros similares.
- Aprendizaje no supervisado: Este algoritmo necesita indicaciones previas, que le ensean a comprender y analizar la informacin, para resolver problemas futuros similares. No neceista datos etiquetados.
- `Aprendizaje de refuerzo`: Este algoritmo aprende por su cuenta, en base a unos conocimientos previamente introducidos y a la prctica que realiza sobre los problemas, aprendiendo en funcin del xito o fracaso que obtiene al resolver los problemas.

## Scikit-learn

### Modo uso
- Importar el mdulo
    `from sklearn.linear_model import LinearRegression`
- Inicializar los parmetros del modelo. Ejemplo:
    `modelo=LinearRegression(normalize=True)`
- Los datos disponibles se dividen en datos de entrenamiento y de prueba. 
    - `train_test_split` es una funcin de scikit-learn que se utiliza para dividir los datos en dos conjuntos: el conjunto de entrenamiento y el conjunto de prueba.
- Entrenar el modelo con los datos de entrenamiento.
    - `modelo.fit(X_train, y_train)`
- Predecir las etiquetas o valores para losd atos.
    - `predicciones = modelo.predict(X_test)`
- Ahora podemos evaluar el rendimiento del modelo. Comparando nuestras predicciones con los valores correctos de los datos de prueba.
