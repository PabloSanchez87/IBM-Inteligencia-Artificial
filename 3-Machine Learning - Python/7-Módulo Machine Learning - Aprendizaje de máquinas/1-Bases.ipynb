# Machine Learning
- Subcampo de las ciencias de la computación y una rama de la inteligencia artificial, cuyo objetivo es desarrollar técnicas que permitan que las computadoras aprendan.
- Es un método de análisis de datos que automatiza la contruscción de un modelo analítico.
- Permite a los ordenadores encontrar soluciones a problemas, sin ser explícitamente programados para ellos, gracias al uso de algoritmos que aprenden a partir de datos.

    

## Tipos de algoritmos de Machine Learning
- `Aprendizaje supervisado`: Este algoritmo necesita datos prviamente etiquetados para aprender a realizar el trabajo. En base a estos datos, el algoritmo es capaz de aprender a resolver problemas futuros similares.
- Aprendizaje no supervisado: Este algoritmo necesita indicaciones previas, que le enseñan a comprender y analizar la información, para resolver problemas futuros similares. No neceista datos etiquetados.
- `Aprendizaje de refuerzo`: Este algoritmo aprende por su cuenta, en base a unos conocimientos previamente introducidos y a la práctica que realiza sobre los problemas, aprendiendo en función del éxito o fracaso que obtiene al resolver los problemas.

## Scikit-learn
- Scikit-learn es una librería de aprendizaje de máquina para Python
    - pip install scikit-learn

### Modo uso
- Importar el módulo
    `from sklearn.linear_model import LinearRegression`
- Inicializar los parámetros del modelo. Ejemplo:
    `modelo=LinearRegression(normalize=True)`
- Los datos disponibles se dividen en datos de entrenamiento y de prueba. 
    - `train_test_split` es una función de scikit-learn que se utiliza para dividir los datos en dos conjuntos: el conjunto de entrenamiento y el conjunto de prueba.
- Entrenar el modelo con los datos de entrenamiento.
    - `modelo.fit(X_train, y_train)`
- Predecir las etiquetas o valores para losd atos.
    - `predicciones = modelo.predict(X_test)`
- Ahora podemos evaluar el rendimiento del modelo. Comparando nuestras predicciones con los valores correctos de los datos de prueba.
